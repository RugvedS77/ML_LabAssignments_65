```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Rugved\Downloads\Heart (1).csv")
```

```python

# Display first few rows
print("First five rows of the dataset:")
display(df.head())

# Shape of the dataset
print("\nShape of the dataset:")
print(df.shape)

# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Data Types
print("\nData types of each column:")
print(df.dtypes)

# Unique Values
print("\nNumber of unique values in each column:")
print(df.nunique())

# Summary Statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
df['Ca'] = df['Ca'].fillna(df['Ca'].mean())
df['Thal'] = df['Thal'].fillna('Unknown')
```

```python
# Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Age'], y=df['Chol'], hue=df['AHD'])
plt.xlabel("Age")
plt.ylabel("Cholesterol Level")
plt.title("Scatter Plot: Age vs Cholesterol")
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(8,5))
sns.countplot(x='ChestPain', data=df, palette='viridis')
plt.title("Bar Plot of Chest Pain Types")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['Age', 'RestBP', 'Chol', 'MaxHR', 'Oldpeak']])
plt.xticks(rotation=45)
plt.title("Box Plot for Numerical Features")
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(6,6))
df['AHD'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90, shadow=True)
plt.title("Pie Chart of Heart Disease (AHD)")
plt.ylabel("")  # Hide y-label
plt.show()
```

```python
# Line Chart
plt.figure(figsize=(8,5))
sns.lineplot(x=df['Age'], y=df['MaxHR'], marker='o', color='blue')
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.title("Line Chart: Age vs MaxHR")
plt.show()
```
